```python
#imported the global economic indicators dataset from kaggle to be used for EDA. 
import pandas as pd
myfile=pd.read_csv('world_bank.csv')
myfile.head()
```

```python
#I am going to use pandas, seaborn(cause it is better than matplotlib) and sklearn for EDA. 
#using pandas first 
# myfile.shape                 # gives Rows and columns
myfile.columns               # gives Column names
# myfile.dtypes                # gives Data types
# myfile.tail()                   #shows end last 5 of the data
# myfile.info()                # gives Non-null values and dtypes
# myfile.describe()            # Summary of numerical columns
```

```python
myfile.isnull().sum() #checks how many empty spaces are there, cause there are many in the file 
myfile.dropna(inplace=True)

print(myfile[myfile["Unemployment Rate (%)"]>10])
```

```python
#checking the relation between higher unemployment rate and lower GDP per capita 
newfile=myfile[(myfile["Unemployment Rate (%)"]>8) & (myfile["GDP per Capita (Current USD)"]>5000)]
print(newfile)
import seaborn as sb   #plotting a graph now 
sb.lineplot(x='Unemployment Rate (%)' , y='GDP per Capita (Current USD)' ,  data=myfile) 
sb.lineplot(x='Unemployment Rate (%)' , y='GDP per Capita (Current USD)' ,  data=newfile)  #now it will draw the graph of filtered conditions and it will also compare both. 
```

```python
#comparision between different conditions of interest rate and inflation
newfile2=myfile[(myfile['Interest Rate (Real, %)']>5) &(myfile['Inflation (GDP Deflator, %)']<-2)]
print(newfile2)
#using matplotlib now and in that, scatter plot 
import matplotlib.pyplot as mp

myfile.plot(kind='scatter' , x='Interest Rate (Real, %)' , y='Inflation (GDP Deflator, %)')
mp.figure()
sb.lineplot(x='Interest Rate (Real, %)' , y='Inflation (GDP Deflator, %)', data=myfile)
mp.title("difference between matplotlib and seaborn")
```

```python
#using sklearn library now 
from sklearn.preprocessing import LabelEncoder

le_country = LabelEncoder()
le_id = LabelEncoder()

myfile['country_name_encoded'] = le_country.fit_transform(myfile['country_name'].astype(str))
myfile['country_id_encoded'] = le_id.fit_transform(myfile['country_id'].astype(str))
print(myfile[['country_name', 'country_name_encoded']].head())
```
