```python
#we have taken the stock prices of the company NVIDEA to get a linear straight line for our training to run good. I had done it on our student dataset also and it worked okayish (the graph was semi linear)
import pandas as pd 
myfile=pd.read_csv('Nvidia.csv')
myfile.head()
```

```python
import matplotlib.pyplot as mb
myfile.plot(kind='scatter' , x='Open' , y='High')
# myfile.plot(kind='scatter' , x='Close' , y='High')
```

```python
X=myfile[['Open']]
Y=myfile['High']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train, Y_train)   #so now our model is trained 
```
